```python
df['months since last claim'].fillna(df['months since last claim'].mean(), inplace=True)
df['number of open complaints'].fillna(df['number of open complaints'].mean(), inplace=True)

```

```python

df['state'].fillna('California', inplace=True)
df['response'].fillna('No', inplace=True)
df['vehicle class'].fillna('Four-Door Car', inplace=True)
df['vehicle size'].fillna('Medsize', inplace=True)
df['vehicle type'].fillna('A', inplace=True)
```

```python
import datetime
from datetime import date
```

```python
#Extract the months from the dataset and store in a separate column

df['effective to date'] = pd.to_datetime(df['effective to date'])
```

```python
df
```

```python
df['effective to date: month'] = df['effective to date'].dt.month
```

```python
df
```

```python
mask_first_quarter = df['effective to date: month'] <= 3
```

```python
display(df[mask_first_quarter])
```

```python
#Bonus

def datetime_format(x):
    for x in df['effective to date']:
        return pd.to_datetime(x)
        
df["effective to date"] = df["effective to date"].apply(lambda x: datetime_format(x))
```

```python
df
```

```python
df['effective to date: month'] = df["effective to date"].apply(lambda x: datetime_format(x).month)
```

```python
df
```

```python
df_month = df["effective to date: month"].apply(lambda x: datetime_format(x).month <= 3)
df[df_month]
```
